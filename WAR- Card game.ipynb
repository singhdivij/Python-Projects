# WAR: CARD GAME

LET'S START BY CREATING A CLASS NAMED CARD THAT HOLDS THE ATTRIBUTES OF CARDS:

1. Suit
2. Rank
3. Value

ALSO, DEFINE A GLOBAL VARIABLE CALLED VALUES THAT STORES A DICTIONARY WHICH HOLDS KEY VALUE PAIRS OF CARD-RANK: AN INTEGER.

AND TWO OTHER GLOBAL VARIABLES FOR SUITS AND RANKS. THEY ARE BASICALLY TUPLES.    


```python
#Global Variable for Rank values
import random
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
```

```python
#Creating the first class called Card

class Card:
    
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
        self.values = values[rank]
        
    def __str__(self):
        return self.rank + ' of ' + self.suit
```

```python
class Deck:
    
    def __init__(self):
        
        self.all_cards = []
        
        for suit in suits:
            for rank in ranks:
                #Create the Card Object
                created_card = Card(suit,rank)
                self.all_cards.append(created_card)
    
    def shuffle(self):
        
        random.shuffle(self.all_cards)
        
    def deal_one(self):
        
        return self.all_cards.pop()
```

```python
class Player:
    
    def __init__(self, name):
        
        self.name = name
        self.all_cards = []
        
    def remove_one(self):
        return self.all_cards.pop(0)
    
    def add_cards(self, new_cards):
        if type(new_cards) == type([]):
            # FOR ADDING MULTIPLE CARD OBJECTS
            return self.all_cards.extend(new_cards)
        else:
            # FOR A SINGLE CARD OBJECT
            self.all_cards.append(new_cards)
    
    def __str__(self):
        return f"Player {self.name} has {len(self.all_cards)} cards"
```

```python
# GAME SETUP
player_one = Player("warrior One")
player_two = Player("Warrior Two")

new_deck = Deck()
new_deck.shuffle()

for cards in range(26):
    player_one.add_cards(new_deck.deal_one())
    player_two.add_cards(new_deck.deal_one())
```

```python
# G A M E  L O G I C
game_on = True
round_num = 0

while game_on:
    
    round_num += 1
    print(f"Round {round_num}")
    
    if len(player_one.all_cards) == 0:
        print("Warrior One is out of cards!!!\nWARRIOR TWO WINS")
        game_on = False
        break
        
    if len(player_two.all_cards) == 0:
        print("Warrior Two is out of cards!!!\nWARRIOR ONE WINS")
        game_on = False
        break
        
    # START A NEW ROUND
    player_one_cards = []
    player_one_cards.append(player_one.remove_one())
    
    player_two_cards = []
    player_two_cards.append(player_two.remove_one())
    
    at_war = True
        
    while at_war:
        # HEIRARCHY OF CARDS IS JUDGED BY VALUES
        if player_one_cards[-1].values > player_two_cards[-1].values:
            player_one.add_cards(player_one_cards)
            player_one.add_cards(player_two_cards)
            at_war = False
            
        elif player_two_cards[-1].values > player_one_cards[-1].values:
            player_two.add_cards(player_two_cards)
            player_two.add_cards(player_one_cards)
            at_war = False
        
        else:
            print("W A R !")
            
            if len(player_one.all_cards) < 5:
                print("Warrior One cannot to declare WAR!")
                print("Warrior Two WINS!!!")
                game_on = False
                break
                
            elif len(player_two.all_cards) < 5:
                print("Warrior Two cannot to declare WAR!")
                print("Warrior One WINS!!!")
                game_on = False
                break
                
            else:
                for num in range(3):
                    player_one_cards.append(player_one.remove_one())
                    player_two_cards.append(player_two.remove_one())
```
