```python
from IPython.display import clear_output
import random

def playboard(board):                                                # WORKS CORRECTLY
    
    clear_output()
    print('   |   |')
    print(' ' + board[7] + ' | ' + board[8] + ' | ' + board[9])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[4] + ' | ' + board[5] + ' | ' + board[6])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' '+ board[1] + ' | ' + board[2] + ' | ' + board[3])
    print('   |   |')
    

def player_input():
    marker = ' '
    
    while not (marker == 'X' or marker == 'O'):
        marker = input('Player 1: Do you want to be X or O? ').upper()

    if marker == 'X':
        return ('X', 'O')
    else:
        return ('O', 'X')

def choose_first():
    if random.randint(0, 1) == 0:
        return 'Player 2'
    else:
        return 'Player 1'

def space_check(board, position):                              # Function to check for empty space
    return board[position] == ' '

def player_choice(board, position, marker):
    
    position = 0
    
    while position not in range(1,10) or not space_check(board, position):
        position = int(input('Choose position {1 - 9}:'))
        
    return position

def place_marker(board, marker, position):
    
    board[position] = marker

def win_check(board, xoro):
    
    return((board[7] == xoro and board[8] == xoro and  board[9] == xoro) or # across first row
           (board[4] == xoro and board[5] == xoro and  board[6] == xoro) or # across second row
           (board[1] == xoro and board[2] == xoro and  board[3] == xoro) or # across last row
           (board[1] == xoro and board[4] == xoro and  board[7] == xoro) or # across first column
           (board[2] == xoro and board[5] == xoro and  board[8] == xoro) or # across second column
           (board[3] == xoro and board[6] == xoro and  board[9] == xoro) or # across last column
           (board[1] == xoro and board[5] == xoro and  board[9] == xoro) or # across diagonally
           (board[3] == xoro and board[5] == xoro and  board[7] == xoro))   # across diagonally
           
def fullboardcheck(board):
    
    for i in range(1,10):
        if space_check(board, i):
            return False
    return True

def replay():
    choice = input('Wanna play again? Y or N:').upper()
    return choice == 'Y'

# WHILE LOOP TO KEEP THE GAME RUNNING
while True:
    
    # PLAY THE GAME
    
    the_board = [' ']* 10
    player1marker, player2marker = player_input()
    
    turn = choose_first()
    print(turn + ' will go first.')
    
    game_start = input('Ready? Y or N:').upper()
    if game_start == 'Y':
        game_on = True
    else:
        game_on = False
        
    while game_on:                                 # WE ASSUME GAME ON IS TRUE
        if turn == 'Player 1':
            playboard(the_board)
            position = player_choice(the_board, turn, player1marker)
            place_marker(the_board, player1marker, position)
            
            # CHECK IF THEY WIN
            if win_check(the_board, player1marker):
                playboard(the_board)
                print('CONGRATULATIONS PLAYER 1! YOU WON.')
                game_on = False                   # YOU CAN USE A BREAK A FUNCTION AS WELL
            else:
                if fullboardcheck(the_board):
                    playboard(the_board)
                    print('THIS ROUND IS A TIE HOMIES!!!')
                    break
                else:
                    turn = 'Player 2'
                    
        else:
            playboard(the_board)
            position = player_choice(the_board, turn, player2marker)
            place_marker(the_board, player2marker, position)
            
            # CHECK IF THEY WIN
            if win_check(the_board, player2marker):
                playboard(the_board)
                print('CONGRATULATIONS PLAYER 2! YOU WON.')
                game_on = False                   # YOU CAN USE A BREAK A FUNCTION AS WELL
            else:
                if fullboardcheck(the_board):
                    playboard(the_board)
                    print('THIS ROUND IS A TIE HOMIES!!!')
                    break
                else:
                    turn = 'Player 1'
            
    # BREAK OUT OF THE GAME
    if not replay():
        break
```
